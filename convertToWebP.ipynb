```python
import os
from PIL import Image
```

```python
# Iterate through all the files in the folder
def convertToWebP(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
            # Open the image file
            img_path = os.path.join(folder_path, filename)
            with Image.open(img_path) as img:
                # Define the new filename with .webp extension
                new_filename = os.path.splitext(filename)[0] + '.webp'
                new_img_path = os.path.join(folder_path, new_filename)
                
                # Save the image in WebP format
                img.save(new_img_path, 'webp')
                
            print(f"Converted {filename} to {new_filename}")
```

```python
# Define the path to the folder containing images
folder_path = ['./public/', './public/aboutUs/Awards/', 
               './public/aboutUs/Certificates/', 
               './public/gallery/elevator-exhibition-2018/', 
               './public/gallery/elevator-seminar-2022/', 
               './public/gallery/technical-seminar-2016/', 
               './public/products/accessories/', 
               './public/products/adco/', 
               './public/products/asian/', 
               './public/products/monarch/', 
               './public/products/tectronics/', 
               ]
for i in folder_path:
    convertToWebP(i)
```
